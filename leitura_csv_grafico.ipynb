Mapa de densidade

```python
%%time
plt.figure(figsize=[12,5], dpi=300)
plt.suptitle("Estrelas no DES DR2", fontsize=14)
plt.subplot(1,2,1)
plt.hist2d(csvdata.ra, csvdata.dec, bins=50)
plt.xlabel("R.A. (deg)", fontsize=14)
plt.ylabel("Dec. (deg)", fontsize=14)
plt.colorbar(label="Densidade de Pontos")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.subplot(1,2,2)
plt.hist2d(csvdata_1.ra, csvdata_1.dec, bins=50)
plt.xlabel("R.A. (deg)", fontsize=14)
plt.ylabel("Dec. (deg)", fontsize=14)
plt.colorbar(label="Densidade de Pontos")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
```

Grafico CMD

```python
new_columns = {"coadd_object_id": "object_id",
               "mag_auto_g_dered": "mag_g",
               "mag_auto_r_dered": "mag_r",
               "mag_auto_i_dered": "mag_i",
               "magerr_auto_g": "err_g", 
               "magerr_auto_r": "err_r", 
               "magerr_auto_i": "err_i"}

csvdata.rename(columns=new_columns, inplace=True)    
```

```python
csvdata.info()
```

```python
csvdata["gmr"] = csvdata.mag_g - csvdata.mag_r
```

```python
csvdata.head()
```

```python
csvdata.query("flags_g < 4  & mag_g != 99. & mag_r != 99. & mag_i != 99. ", inplace=True)
```

```python
csvdata.count()
```

```python
plt.figure(figsize=[7,5])
plt.title("CMD - Estrelas no DES DR2", fontsize=14)
plt.hexbin(csvdata.gmr, csvdata.mag_r, gridsize=500, bins='log') #Essa linha cria um grÃ¡fico hexbin usando Matplotlib, 
#que Ã© particularmente Ãºtil para visualizar a densidade de pontos de dados em um espaÃ§o bidimensional.
plt.xlabel("$g - r$", fontsize=14)
plt.ylabel("$g$", fontsize=14)
plt.colorbar(label="Densidade de Pontos")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(-1.5,2.5)
plt.ylim(30,14)
plt.tight_layout()
```
